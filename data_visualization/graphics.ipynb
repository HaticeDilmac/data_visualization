```python
import seaborn as sns
diamonds= sns.load_dataset('diamonds')
df =diamonds.copy()
df.head()
```

```python
#displot değişkeni tabloyu oluşturuyor.
#kde argumanınn gfrafikte görünüp görünmemesi durumu.
sns.displot(df.price,kde=False)
```

```python
df["price"].describe()#değerlerin miktarlarını göstermektedir.
```

```python
sns.displot(df.price,bins=10,kde=False) #bins barın kalınlığını belirtir.
```

```python
sns.displot(df.price )
```

```python
sns.kdeplot(df.price, shade=True)
```

```python
(sns
.FacetGrid(df,
             hue="cut",
             height =5,
             xlim = (0,10000))
.map(sns.kdlopt, "price",shade=True)
.add_legend()
)
```
